```python
import pandas as pd
import wget
from zipfile import ZipFile
import plotly.graph_objects as go
```

```python
URL_BASE = 'http://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/DFP/DADOS/'
```

```python
ARQUIVOS_BAIXADOS = list()

for ano in range(2010, 2021):
    ARQUIVOS_BAIXADOS.append(f'dfp_cia_aberta_{ano}.zip')
```

```python
for nomes in ARQUIVOS_BAIXADOS:
    wget.download(URL_BASE+nomes)
```

```python
for arq in ARQUIVOS_BAIXADOS:
    ZipFile(arq, 'r').extractall('CVM_dados')
```

```python
nomes = ['BPA_con', 'BPA_ind', 'BPP_con', 'BPP_ind', 'DFC_MD_con', 'DFC_MD_ind', 'DFC_MI_con', 'DFC_MI_ind', 'DMPL_con', 'DMPL_ind', 'DRE_con', 'DRE_ind', 'DVA_con', 'DVA_ind']
for nome in nomes:
    arquivo = pd.DataFrame()
    for ano in range(2010,2021):
        arquivo = pd.concat([arquivo, pd.read_csv(f'CVM_dados//dfp_cia_aberta_{nome}_{ano}.csv', sep=';', decimal=',', encoding='ISO-8859-1')])
    arquivo.to_csv(f'dados_prontos/dfp_cia_aberta_{nome}_2010-2020.csv', index=False)
```
